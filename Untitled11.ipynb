```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
stockmarket_data = pd.Series(np.random.randn(1000))
```

```python
stockmarket_cumulative = stockmarket_data.cumsum()
```

```python
stockmarket_cumulative.plot()
```
